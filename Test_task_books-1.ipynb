```python
data = pd.read_csv('C:/Users/Tania/Desktop/AWS/test_task/best-selling-books export 2025-01-22 19-29-09.csv')
print(data.head())
```

```python
print(data.isnull().sum())

# Drop rows with missing 'Genre' values
data = data.dropna(subset=['Genre'])
```

```python
genre_counts = data['Genre'].value_counts()

# Display the genre counts
print(genre_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
genre_counts = genre_counts.sort_values(ascending=False)

# Отображаем только топ-10 популярных жанров
top_genres = genre_counts.head(10)

# Построение графика
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='coolwarm')

# Добавление заголовка и подписей
plt.title('Top 10 Most Popular Book Genres', fontsize=16)
plt.xlabel('Number of Books', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Показываем график
plt.tight_layout()
plt.show()
```
